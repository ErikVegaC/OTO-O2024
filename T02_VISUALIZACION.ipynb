```python
# Grupo 1 de funciones :

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10,10,1000)

y1 = 2*x + 1

y2 = -x + 4

y3 = x**2

plt.figure(figsize=(15,10))
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.grid()


```

```python
def funcionT02(x1,x2,n,a,b,k):
  x= np.linspace(x1,x2,n)
  for i in range(k):
    y = np.sin((((b-a)/(k-1))* a + i)*x)
    plt.plot(y,label="a = " +str((((b-a)/(k-1))* a + i)))
    plt.grid()
    plt.title("Funciones - y = sin(ax)")
    plt.xlabel("Eje X")
    plt.ylabel("Eje Y")
    plt.legend()





funcionT02(0,2*np.pi,1000,1,3,3)
```
