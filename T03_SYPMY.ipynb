##### 1. El perímetro de un triángulo isósceles es de 20 cm. Imagina que este triángulo rota centrado en su eje de simetría para formar la ilusión de un cono.

image.png

##### ¿Cuánto deben medir sus lados para que el volumen generado por la rotación sea el mayor posible?

Ecuaciones

 - $$ 2L + B - 20  $$ \\
 - $$ \frac{π}{3} * \frac{B}{2} * \sqrt{L^2 - \frac{B}{4}^2}   $$

```python

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

sp.var("L")
sp.var("B")

f1_1 =  2 * L + B - 20

f2_1 =  (sp.pi*((0.5*B)**2)*(sp.sqrt(((L)**2)-((0.5*B)**2))))/3

solveB_f1_1 = sp.solve(f1_1,B)[0]
subsB_f2_1 = f2_1.subs(B,solveB_f1_1)
diffsubsB_f2_1 = sp.diff(subsB_f2_1,L)

solveL_f2_1 = sp.solve(diffsubsB_f2_1,L)[0]

print(f"Ambos lados de el triángulo isóceles deben medir {int(solveL_f2_1)} cm para que el volumen generado por la rotación sea la máxima posible")
```

2. Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol en un parque. Es una portería simple que tendrá sólo dos postes y el travesaño

image-03630135-c812-4d3e-a96e-c5213162c8be.png

 Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

Ecuaciones

- $$ 2 P + T - 6 $$

- $$ P * T  $$

```python
import matplotlib.pyplot as plt
import sympy as sp

sp.var("P")
sp.var("T")

f1_2 = 2 * P + T - 6

f2_2 = P * T

solveT_f1_2 = sp.solve(f1_2,T)[0]
subsT_f2_2 = f2_2.subs(T,solveT_f1_2)

diffsubsT_f2_2 = sp.diff(subsT_f2_2,P)

solveP = sp.solve(diffsubsT_f2_2,P)[0]
solveT = sp.solve(f1_2.subs(P,solveP),T)[0]

print(f"Para que el área de la porteria sea máxima, los postes deben de medir {float(solveP)} mts y el travesaño {int(solveT)} mts ")
```

3. **Resistencia de una viga**  Una viga de madera tiene una sección transversal rectangular de altura $h$ y ancho $w$. La resistencia $S$ de la viga es directamente proporcional al ancho y al cuadrado de su altura ($S=kwh^2$, donde $k$ es una constante de proporcionalidad). ¿Cuáles son las dimensiones de la viga más fuerte que puede cortarse a partir de un leño redondo de $24$ pulgadas de diámetro?


```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

C = patches.Circle((0, 0),12, fill=False,color='brown',lw=4)
S = patches.Rectangle((-9,-8),18,16,fill=False,color='k')
ax = plt.gca()
ax.add_artist(S)
ax.add_artist(C)

plt.text(0,8.5,'$w$')
plt.text(9.5,0,'$h$')

plt.plot([-9,9],[-8,8],'r--')
plt.text(0,2,'$24$')

plt.axis('equal')
plt.ylim([-15,15])
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import sympy as sp

sp.var("H")
sp.var("W")
sp.var("K")



f1_3 = K * W * (H**2)
f2_3 = (W**2) + (H**2) - (24**2)

solveW_f2_3 = sp.solve(f2_3,W)[1]
subsW_f1_3 = f1_3.subs(W,solveW_f2_3)
diffsubsW_f1_3 = sp.diff(subsW_f1_3,H)
solveH = sp.solve(diffsubsW_f1_3,H)[2]
solveW = sp.solve(f2_3.subs(H,solveH),W)[1]

print(f"Las dimensiones de la viga que se puede obtener de un leño redondo de 24 pulgadas de diámetro son W = {float(solveW)} pulgadas y H = {float(solveH)} pulgadas")

```

5. **Fuerza mínima**  Se diseña un componente para deslizar un bloque de acero con peso $W=10 N$ a través de una mesa y sobre una canaleta. Se opone al movimiento del bloque una fuerza de fricción proporcional a su peso aparente (sea $k=0.1$ la constante de proporcionalidad). Determinar la fuerza mínima $f$ necesaria para deslizar el bloque y encontrar el valor correspondiente de $\theta$.

Ayudín: $f\cos \theta$ es la fuerza de dirección del movimiento, y $f\sin\theta$ es la fuerza que tiende a levantar el bloque. De tal modo, el peso aparente del bloque es $W-f\sin\theta$.

```python
W = patches.Rectangle((0, 2),0.8, 0.5, fill=False,color='k')
a = patches.Arrow(0.8,2.25,1,0.5,width=0.1,color='k')
arc = patches.Arc((0.8,2.25),1,1,theta2=30)
ax = plt.gca()
ax.add_artist(W)
ax.add_artist(a)
ax.add_artist(arc)

plt.text(0.2,2.6,'$W$')
plt.text(1.8,3,'$f$')
plt.text(1.35,2.305,r'$\theta$')

plt.plot([1.8,2.8],[2.75,3.25],'k')
plt.plot([0.8,2],[2.25,2.25],'k--')
plt.plot([-1,-1,2,2],[0,2,2,0],'k')


plt.axis('equal')
plt.ylim([0,4])
plt.axis('off')
plt.show()
```
