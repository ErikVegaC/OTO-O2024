- Calcula la Utilidad (Ingreso - Gasto)

```python
datos["Utilidad"] = datos["Ingreso"] - datos["Gasto"]
```

```python
datos
```

- Promedio de "Gasto" en donde "Utilidad" fue negativa

```python
ut_neg = datos[datos["Utilidad"] < 0 ]
ut_neg
```

```python
ut_neg["Gasto"].mean()
```
